```python
import datetime
import pandas as pd
```

```python
#Mar 20

import snscrape.modules.twitter as sntwitter
import pandas
from datetime import datetime

# Creating list to append tweet data to
tweets_mar20 = []
start_time = datetime.now()

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('(covid OR covid-19 OR coronavirus OR pandemic OR lockdown) AND (depression OR depressed OR anxiety) since:2020-03-01 until:2020-04-01').get_items()):
    if i>500:
        break
    tweets_mar20.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.lang])
    
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
   

```

```python
df_merged=pd.concat([pd.DataFrame(d) for d in tweet_list])
df_merged.columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Language']
```

```python
df_merged
```

```python
df_merged['Date'] = df_merged['Datetime'].dt.date
df_merged['Month'] = df_merged['Datetime'].dt.month
df_merged['Year'] = df_merged['Datetime'].dt.year
```

```python
df_merged
```

```python
import calendar
df_merged['Month'] = df_merged['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df_merged.Year.value_counts()
```

```python
df_merged=df_merged.sort_values('Date')
```

```python
df_merged.reset_index(inplace = True)
```

```python
df_merged['Label']=1
df_merged
```

```python
df_merged.Language.value_counts()
```

```python
df_merged=df_merged[df_merged['Language']=='en']
df_merged
```

```python
df_merged.isnull().sum()
```

```python
df_merged[df_merged['Tweet Id'].duplicated(keep=False)].sort_values("Tweet Id")
```

```python
df_merged.drop_duplicates( "Tweet Id" , keep='first', inplace=True)
```

```python
df_merged.reset_index(inplace = True)
```

```python
df_merged
```

```python
df_merged.drop(columns=['level_0','index','Datetime','Language','Date'], inplace=True)
```

```python
df_merged=df_merged[['Label','Tweet Id', 'Username', 'Text','Year','Month']]
```
